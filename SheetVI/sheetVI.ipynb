La confianza de la regla $( X \rightarrow Y )$ se define como:

$\text{conf}(X \rightarrow Y) = \frac{\text{sup}(XY)}{\text{sup}(X)}$

Para la regla $( W \rightarrow (Z \setminus W) )$, la confianza se define como:

$ \text{conf}(W \rightarrow (Z \setminus W)) = \frac{\text{sup}(ZW)}{\text{sup}(W)} $

Dado que $( W \subset X )$ y $( Z \supset XY )$, podemos escribir $( Z = (Z \setminus W) \cup W )$. Entonces, la confianza de $( W \rightarrow (Z \setminus W) )$ se puede reescribir como:

$ \text{conf}(W \rightarrow (Z \setminus W)) = \frac{\text{sup}(ZW)}{\text{sup}(W)} = \frac{\text{sup}(XY \cup W)}{\text{sup}(W)} $

Usando la propiedad de subconjunto de $( W )$ en $( X )$, podemos decir que $( \text{sup}(XY \cup W) \geq \text{sup}(XY) )$, ya que $( XY \cup W = X )$. Por lo tanto, $\\( \text{conf}(W \rightarrow (Z \setminus W)) )$ es menor o igual a $( \text{conf}(X \rightarrow Y) )$, lo que se demuestra:

$ \text{conf}(W \rightarrow (Z \setminus W)) \leq \text{conf}(X \rightarrow Y) $
